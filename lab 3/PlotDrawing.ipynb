```python
import matplotlib.pyplot as plt
```

```python
# 
epochs = list(range(1, 11))
lenet_accuracy = [88.25, 96.88, 96.54, 98.14, 98.10, 98.50, 98.46, 88.44, 98.52, 98.41]
alexnet_accuracy = [86.06, 98.33, 98.76, 98.88, 99.00, 99.09, 99.07, 99.32, 99.24, 99.27]
resnet_accuracy = [95.24, 98.96, 99.15, 99.24, 89.83, 99.24, 99.30, 99.19, 99.26, 99.48]
googlenet_accuracy = [89.82, 94.09, 95.63, 96.10, 96.82, 97.07, 96.79, 97.17, 96.92, 97.47]

# 
plt.figure(figsize=(10, 6))

plt.plot(epochs, lenet_accuracy, label='LeNet', marker='o')
plt.plot(epochs, alexnet_accuracy, label='AlexNet', marker='o')
plt.plot(epochs, resnet_accuracy, label='ResNet', marker='o')
plt.plot(epochs, googlenet_accuracy, label='GoogleNet', marker='o')

# 
plt.title('Validation Accuracy - Epoch')
plt.xlabel('Epoch')
plt.ylabel('Validation Accuracy (%)')
plt.ylim(85, 100)
plt.xticks(epochs)
plt.legend()
plt.grid(True)

# 
plt.show()
```

```python
# 
models = ['LeNet', 'AlexNet', 'ResNet', 'GoogleNet']
test_accuracy = [98.50, 99.33, 99.52, 98.05]

# 
plt.figure(figsize=(8, 5))
plt.bar(models, test_accuracy, color=['blue', 'green', 'orange', 'red'])

# 
plt.title('Test Accuracy - Model')
plt.xlabel('Model')
plt.ylabel('Test Accutacy (%)')
plt.ylim(95, 100)  

# 
for i, v in enumerate(test_accuracy):
    plt.text(i, v + 0.2, f'{v:.2f}%', ha='center', va='bottom')

# 
plt.show()
```
